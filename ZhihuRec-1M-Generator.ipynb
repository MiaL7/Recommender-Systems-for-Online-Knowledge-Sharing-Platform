```python
import pandas as pd
import numpy as np
import os
```

```python
# stat for generating ZhihuRec-1M
IMPRESSION = 999970
QUERIES = 38422
USERS = 7974
ANSWERS = 81563
QUESTIONS = 29340
AUTHORS = 47888
TOPICS = 22897
TOKENS = 249586

DIR = "/content/drive/Shareddrives/SI650_Final_Project/ZhihuRec-1M"
SOURCE_DIR = "/content/drive/Shareddrives/SI650_Final_Project/zhihurec"
```

```python
if not os.path.exists(DIR):
    os.mkdir(DIR)
```

```python
impression = pd.read_csv(os.path.join(SOURCE_DIR, "inter_impression.csv"), header=None)
impression.columns = ["userID", "answerID", "imp_ts", "click_ts"]
impression = impression.head(IMPRESSION)
impression.to_csv(os.path.join(DIR, "impression.csv"))
del impression

queries = pd.read_csv(os.path.join(SOURCE_DIR, "inter_query.csv"), header=None)
queries.columns = ["userID", "tokenID", "query_ts"]
queries = queries.head(QUERIES)
queries.to_csv(os.path.join(DIR, "query.csv"))
del queries

users = pd.read_csv(os.path.join(SOURCE_DIR, "info_user.csv"), header=None)
users.columns = [
    "userID", 
    "reg_ts", 
    "gender", 
    "login_freq", 
    "n_followers", 
    "n_topics_followed", 
    "n_questions_followed",
    "n_answers",
    "n_questions",
    "n_comments",
    "n_thanks_recv",
    "n_comments_recv",
    "n_likes_recv",
    "n_dislikes_recv",
    "reg_type",
    "reg_platform",
    "is_android",
    "is_iphone",
    "is_ipad",
    "is_pc",
    "is_mobile",
    "device_model",
    "device_brand",
    "platform",
    "province",
    "city",
    "topicID_followed",
]
users = users.head(USERS)
users.to_csv(os.path.join(DIR, "user.csv"))
del users
```

```python
answer = pd.read_csv(os.path.join(SOURCE_DIR, "info_answer.csv"), header=None)
answer.columns = [
    "answerID",
    "questionID",
    "is_anonymous",
    "authorID",
    "is_hi_val",
    "is_editor_recom",
    "creat_ts",
    "has_pic",
    "has_vid",
    "n_thanks",
    "n_likes",
    "n_comments",
    "n_collections",
    "n_dislikes",
    "n_reports",
    "n_helpless",
    "tokenID",
    "topicID"
]
answer = answer.head(ANSWERS)
answer.to_csv(os.path.join(DIR, "answer.csv"))
del answer

question = pd.read_csv(os.path.join(SOURCE_DIR, "info_question.csv"), header=None)
question.columns = [
    "questionID",
    "create_ts",
    "n_answers",
    "n_followers",
    "n_invitations",
    "n_comments",
    "tokenID",
    "topicID",
]
question = question.head(QUESTIONS)
question.to_csv(os.path.join(DIR, "question.csv"))
del question

author = pd.read_csv(os.path.join(SOURCE_DIR, "info_author.csv"), header=None)
author.columns = [
    "authorID",
    "is_exec_author",
    "n_followers",
    "is_exec_answerer",
]
author = author.head(AUTHORS)
author.to_csv(os.path.join(DIR, "author.csv"))
del author

topic = pd.read_csv(os.path.join(SOURCE_DIR, "info_topic.csv"), header=None)
topic.columns = ["topicID"]
topic = topic.head(TOPICS)
topic.to_csv(os.path.join(DIR, "topic.csv"))
del topic

token = pd.read_csv(os.path.join(SOURCE_DIR, "info_token.csv"), header=None)
token.columns = [
    "tokenID",
    "word_vec",
]
token = token.head(TOKENS)
token.to_csv(os.path.join(DIR, "token.csv"))
del token
```
