```python
import os
import pandas as pd
import numpy as np
users = pd.read_csv(os.path.join(SOURCE_DIR, "user.csv"))
```

```python
users.columns
```

```python
n_users = np.max(users.userID)
n_users
```

```python
# user1, user2, #common_followed_topics
net = []

user2topics = {}

for _, row in users.iterrows():
    if isinstance(row['topicID_followed'], str):
        user2topics[row['userID']] = set(map(int, row['topicID_followed'].split()))
```

```python
from tqdm.notebook import tqdm
```

```python
for user1 in tqdm(user2topics.keys()):
    for user2 in range(user1, n_users+1):
        if user2 in user2topics:
            common = user2topics[user1].intersection(user2topics[user2])
            if len(common) == 0:
                continue
            net.append([user1, user2, len(common)])
```

```python
net_df = pd.DataFrame(net, index=None, columns=["user1:token", "user2:token", "n_common_topics:token"])
```
