## Inter

```python
test_df = pd.read_csv(os.path.join(SOURCE_DIR, "test-1M.csv"), index_col=0)
val_df = pd.read_csv(os.path.join(SOURCE_DIR, "val-1M.csv"), index_col=0)
train_df = pd.read_csv(os.path.join(SOURCE_DIR, "train-1M.csv"), index_col=0)
```

```python
val_df
```

```python

def modify_col_name(df):
  # modify the columns' names
  df = df[['userID', 'answerID', 'is_clicked', 'imp_ts']]
  df['is_clicked'] = df['is_clicked'].apply(lambda x: 1.0 if x else 0.0)
  cols = list(df.columns)
  cols = [col + ":token" for col in cols]
  df.columns = cols
  return df
  
```

```python
test_df = modify_col_name(test_df)
val_df = modify_col_name(val_df)
train_df = modify_col_name(train_df)
```

```python
train_df.to_csv(os.path.join(DIR, "ZhihuRec-1M.train.inter"), index=False)
val_df.to_csv(os.path.join(DIR, "ZhihuRec-1M.val.inter"), index=False)
test_df.to_csv(os.path.join(DIR, "ZhihuRec-1M.test.inter"), index=False)
```

```python
impr = pd.read_csv(os.path.join(SOURCE_DIR, "impression.csv"), index_col=0)
impr["is_clicked"] = impr["click_ts"] != 0
impr["is_clicked"] = impr["is_clicked"].apply(lambda x: 1 if x else 0)
impr
```

```python
# modify the columns' names
impr_cols = list(impr.columns)
impr_cols = [col + ":token" for col in impr_cols]
impr.columns = impr_cols
impr.columns
```

```python
TRAIN_RATIO = 0.9
y_train = impr.sort_values("imp_ts:token").iloc[:int(impr.shape[0]*TRAIN_RATIO), :]
y_test = impr.sort_values("imp_ts:token").iloc[int(impr.shape[0]*TRAIN_RATIO):, :]
```

```python
ts_cutoff = y_train.iloc[-1, :]['imp_ts:token']
ts_cutoff
```

```python
impr_train = impr[impr['imp_ts:token'] < ts_cutoff]
impr_test = impr[impr['imp_ts:token'] >= ts_cutoff]
```

```python
from sklearn.model_selection import train_test_split

impr_train, impr_val = train_test_split(impr_train, test_size=0.2)
```

```python
impr_train.to_csv(os.path.join(DIR, "ZhihuRec-1M.train.inter"), index=False)
impr_val.to_csv(os.path.join(DIR, "ZhihuRec-1M.val.inter"), index=False)
impr_test.to_csv(os.path.join(DIR, "ZhihuRec-1M.test.inter"), index=False)
```
